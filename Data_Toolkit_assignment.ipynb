1. What is NumPy, and why is it widely used in Python?

2. How does broadcasting work in NumPy?
- Broadcasting is NumPy's ability to perform element-wise operations on arrays with different shapes without explicitly reshaping them. It follows specific rules:

- Arrays are aligned from the rightmost dimension
Dimensions of size 1 can be "stretched" to match larger dimensions
Missing dimensions are assumed to be of size 1

- Example: Adding a scalar to an array, or adding arrays of shapes (3,1) and (1,4) results in (3,4).

```python
#4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create sample dataset
np.random.seed(42)
data = {
    'Height': np.random.normal(170, 10, 100),
    'Weight': np.random.normal(70, 15, 100),
    'Age': np.random.randint(18, 65, 100),
    'Income': np.random.normal(50000, 20000, 100)
}

# Add some correlation
data['Weight'] = data['Height'] * 0.8 + np.random.normal(0, 5, 100)
data['Income'] = data['Age'] * 800 + np.random.normal(0, 10000, 100)

df = pd.DataFrame(data)

# Calculate correlation matrix
correlation_matrix = df.corr()

# Create heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix,
            annot=True,  # Show correlation values
            cmap='coolwarm',  # Color scheme
            center=0,  # Center colormap at 0
            square=True,  # Make cells square
            fmt='.2f')  # Format numbers to 2 decimal places

plt.title('Correlation Matrix Heatmap')
plt.tight_layout()
plt.show()
```

```python
#6. Create a DataFrame and add a new column based on an existing column.

import pandas as pd

# Create initial DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
        'Age': [25, 30, 35, 28, 22],
        'Salary': [50000, 60000, 70000, 55000, 45000]}

df = pd.DataFrame(data)

# Method 1: Simple calculation
df['Annual_Bonus'] = df['Salary'] * 0.1

# Method 2: Conditional logic using np.where
import numpy as np
df['Age_Group'] = np.where(df['Age'] >= 30, 'Senior', 'Junior')

# Method 3: Using apply with lambda function
df['Salary_Grade'] = df['Salary'].apply(lambda x: 'High' if x >= 60000
                                       else 'Medium' if x >= 50000
                                       else 'Low')

# Method 4: Complex calculation combining multiple columns
df['Salary_per_Age'] = df['Salary'] / df['Age']

print("Updated DataFrame:")
print(df)
```

```python
#7. Write a program to perform element-wise multiplication of two NumPy arrays.

import numpy as np

# Create two NumPy arrays
array1 = np.array([1, 2, 3, 4, 5])
array2 = np.array([2, 3, 4, 5, 6])

# Element-wise multiplication
result = array1 * array2
# Alternative: result = np.multiply(array1, array2)

print("Array 1:", array1)
print("Array 2:", array2)
print("Element-wise multiplication:", result)

# For 2D arrays
array_2d_1 = np.array([[1, 2], [3, 4]])
array_2d_2 = np.array([[5, 6], [7, 8]])

result_2d = array_2d_1 * array_2d_2

print("\n2D Array 1:")
print(array_2d_1)
print("2D Array 2:")
print(array_2d_2)
print("2D Element-wise multiplication:")
print(result_2d)

# Broadcasting example
array_broadcast = np.array([1, 2, 3])
scalar = 5
result_broadcast = array_broadcast * scalar

print(f"\nArray: {array_broadcast}")
print(f"Scalar: {scalar}")
print(f"Broadcast multiplication: {result_broadcast}")
```

```python
#8. Create a line plot with multiple lines using Matplotlib.

import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

# Create line plot with multiple lines
plt.figure(figsize=(10, 6))

plt.plot(x, y1, label='sin(x)', linewidth=2, color='blue')
plt.plot(x, y2, label='cos(x)', linewidth=2, color='red', linestyle='--')
plt.plot(x, y3, label='sin(x)*cos(x)', linewidth=2, color='green', linestyle=':')

plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Multiple Line Plot')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Alternative method using subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# First subplot - all lines together
ax1.plot(x, y1, label='sin(x)', linewidth=2)
ax1.plot(x, y2, label='cos(x)', linewidth=2)
ax1.plot(x, y3, label='sin(x)*cos(x)', linewidth=2)
ax1.set_title('All Lines Together')
ax1.legend()
ax1.grid(True)

# Second subplot - separate y-axes
ax2_twin = ax2.twinx()
line1 = ax2.plot(x, y1, 'b-', label='sin(x)')
line2 = ax2_twin.plot(x, y2, 'r--', label='cos(x)')
ax2.set_title('Separate Y-axes')
ax2.set_ylabel('sin(x)', color='b')
ax2_twin.set_ylabel('cos(x)', color='r')

plt.tight_layout()
plt.show()
```

```python
#9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.

import pandas as pd
import numpy as np

# Generate sample DataFrame
np.random.seed(42)
data = {
    'Name': [f'Person_{i}' for i in range(1, 21)],
    'Age': np.random.randint(18, 65, 20),
    'Salary': np.random.randint(30000, 100000, 20),
    'Department': np.random.choice(['IT', 'HR', 'Finance', 'Marketing'], 20)
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df.head(10))

# Filter rows where Salary > 50000
threshold = 50000
filtered_df = df[df['Salary'] > threshold]

print(f"\nFiltered DataFrame (Salary > {threshold}):")
print(filtered_df)

# Multiple conditions
high_salary_young = df[(df['Salary'] > 60000) & (df['Age'] < 40)]
print(f"\nHigh salary and young employees:")
print(high_salary_young)

# Using query method (alternative approach)
filtered_query = df.query('Salary > 50000 and Age < 50')
print(f"\nUsing query method:")
print(filtered_query)

# Filter with isin for categorical data
it_finance = df[df['Department'].isin(['IT', 'Finance'])]
print(f"\nIT and Finance departments:")
print(it_finance)
```

```python
#10. Create a histogram using Seaborn to visualize a distribution.

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate sample data
np.random.seed(42)
data = np.random.normal(100, 15, 1000)  # Normal distribution

# Basic histogram
plt.figure(figsize=(12, 8))

# Subplot 1: Basic histogram
plt.subplot(2, 2, 1)
sns.histplot(data, bins=30, kde=False)
plt.title('Basic Histogram')

# Subplot 2: Histogram with KDE
plt.subplot(2, 2, 2)
sns.histplot(data, bins=30, kde=True)
plt.title('Histogram with KDE')

# Subplot 3: Multiple distributions
data2 = np.random.normal(120, 20, 1000)
df = pd.DataFrame({'Group_A': data, 'Group_B': data2})
df_melted = df.melt(var_name='Group', value_name='Value')

plt.subplot(2, 2, 3)
sns.histplot(data=df_melted, x='Value', hue='Group', alpha=0.7)
plt.title('Multiple Distributions')

# Subplot 4: Distribution plot (alternative)
plt.subplot(2, 2, 4)
sns.distplot(data, bins=30, hist=True, kde=True, rug=True)
plt.title('Distribution Plot with Rug')

plt.tight_layout()
plt.show()

# Advanced: Faceted histograms
tips = sns.load_dataset('tips')
plt.figure(figsize=(12, 6))
sns.histplot(data=tips, x='total_bill', hue='time', multiple='stack')
plt.title('Restaurant Bills by Time of Day')
plt.show()
```

```python
#11. Perform matrix multiplication using NumPy.

import numpy as np

# Create matrices
matrix_a = np.array([[1, 2, 3],
                     [4, 5, 6]])

matrix_b = np.array([[7, 8],
                     [9, 10],
                     [11, 12]])

# Method 1: Using np.dot()
result_dot = np.dot(matrix_a, matrix_b)

# Method 2: Using @ operator (Python 3.5+)
result_at = matrix_a @ matrix_b

# Method 3: Using np.matmul()
result_matmul = np.matmul(matrix_a, matrix_b)

print("Matrix A (2x3):")
print(matrix_a)
print("\nMatrix B (3x2):")
print(matrix_b)
print("\nMatrix multiplication result (2x2):")
print(result_dot)

# Element-wise multiplication (different from matrix multiplication)
matrix_c = np.array([[1, 2],
                     [3, 4]])
matrix_d = np.array([[5, 6],
                     [7, 8]])

element_wise = matrix_c * matrix_d  # Element-wise
matrix_mult = matrix_c @ matrix_d   # Matrix multiplication

print("\nMatrix C:")
print(matrix_c)
print("Matrix D:")
print(matrix_d)
print("Element-wise multiplication:")
print(element_wise)
print("Matrix multiplication:")
print(matrix_mult)

# Higher dimensional arrays
array_3d_a = np.random.rand(2, 3, 4)
array_3d_b = np.random.rand(2, 4, 5)
result_3d = array_3d_a @ array_3d_b

print(f"\n3D array multiplication:")
print(f"Shape A: {array_3d_a.shape}")
print(f"Shape B: {array_3d_b.shape}")
print(f"Result shape: {result_3d.shape}")
```
